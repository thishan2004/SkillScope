```python
import pandas as pd

postings = pd.read_csv("../data/raw/postings.csv")

print(postings.shape)
```

```python
data_science_keywords = [
    "Data Scientist",
    "Data Analyst",
    "Data Engineer",
    "Machine Learning Engineer",
    "BI Analyst",
    "Business Intelligence",
    "AI Engineer"
]
```

```python
pattern = "|".join(data_science_keywords)

ds_jobs = postings[
    postings["title"]
    .str.contains(pattern, case=False, na=False)
]

print(ds_jobs.shape)
```

```python
ds_jobs["title"].value_counts().head(20)
```

```python
ds_jobs.to_csv(
    "../data/processed/ds_jobs.csv",
    index=False
)

print("Dataset Saved!")
```

```python
import pandas as pd
ds_jobs = pd.read_csv("../data/processed/ds_jobs.csv")
skill_df = pd.read_csv("../data/processed/skill_demand.csv")

print("Datasets loaded successfully!")
print("DS Jobs:", ds_jobs.shape)
print("Skill Demand:", skill_df.shape)
```

```python
skill_df.head(15)
```

```python
top_market_skills = skill_df.head(10)["Skill"].tolist()

top_market_skills
```

```python
my_skills = [
    "Python",
    "SQL",
    "Pandas",
    "NumPy",
    "Machine Learning",
    "Git"
]
```

```python
matched_skills = []

for skill in top_market_skills:
    if skill in my_skills:
        matched_skills.append(skill)

missing_skills = []

for skill in top_market_skills:
    if skill not in my_skills:
        missing_skills.append(skill)

print("Matched Skills:", matched_skills)
print("Missing Skills:", missing_skills)
```

```python
employability_score = (
    len(matched_skills) / len(top_market_skills)
) * 100

print(f"Employability Score: {employability_score:.2f}/100")
```

```python
print("===== SkillScope Employability Report =====")
print(f"Employability Score: {employability_score:.2f}/100")

print("\nSkills You Have:")
for skill in matched_skills:
    print("✅", skill)

print("\nSkills You Need to Improve:")
for skill in missing_skills:
    print("❌", skill)
```

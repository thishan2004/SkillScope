```python
import pandas as pd
ds_jobs = pd.read_csv("../data/processed/ds_jobs.csv")
skill_df = pd.read_csv("../data/processed/skill_demand.csv")

print("Datasets loaded successfully!")
print("DS Jobs:", ds_jobs.shape)
print("Skill Demand:", skill_df.shape)
```

```python
top_market_skills = skill_df.head(10)["Skill"].tolist()

top_market_skills
```

```python
def calculate_employability_score(user_skills, market_skills):
    matched_skills = []
    missing_skills = []

    for skill in market_skills:
        if skill in user_skills:
            matched_skills.append(skill)
        else:
            missing_skills.append(skill)

    score = (len(matched_skills) / len(market_skills)) * 100

    return round(score, 2), matched_skills, missing_skills
```

```python
def recommend_career(user_skills):
    career_paths = {
        "Data Analyst": ["SQL", "Excel", "Power BI", "Tableau", "Statistics"],
        "Data Scientist": ["Python", "SQL", "Machine Learning", "Statistics", "Pandas"],
        "Data Engineer": ["SQL", "Python", "AWS", "Spark", "ETL"],
        "Machine Learning Engineer": ["Python", "Machine Learning", "Deep Learning", "TensorFlow", "PyTorch"],
        "BI Analyst": ["SQL", "Power BI", "Tableau", "Excel", "Statistics"]
    }

    recommendations = []

    for role, required_skills in career_paths.items():
        matched = []
        missing = []

        for skill in required_skills:
            if skill in user_skills:
                matched.append(skill)
            else:
                missing.append(skill)

        score = (len(matched) / len(required_skills)) * 100

        recommendations.append({
            "Role": role,
            "Match Score": round(score, 2),
            "Matched Skills": matched,
            "Missing Skills": missing
        })

    recommendation_df = pd.DataFrame(recommendations)
    recommendation_df = recommendation_df.sort_values(
        by="Match Score",
        ascending=False
    )

    return recommendation_df
```

```python
user_skills = ["Python", "SQL", "Machine Learning"]

score, matched, missing = calculate_employability_score(
    user_skills,
    top_market_skills
)

print("Employability Score:", score)
print("Matched Skills:", matched)
print("Missing Skills:", missing)
```

```python
recommendation_df = recommend_career(user_skills)

recommendation_df
```

```python
best_role = recommendation_df.iloc[0]

print("Recommended Career:", best_role["Role"])
print("Match Score:", best_role["Match Score"])
print("Matched Skills:", best_role["Matched Skills"])
print("Missing Skills:", best_role["Missing Skills"])
```
